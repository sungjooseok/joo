```python
# 네트워크 구조 만들기
 
 ### 1. 기본적인 네트워크 구조 만들기


import networkx as nx # networkx 를 nx로 이용할 수 있게 불러오기
import matplotlib.pyplot as plt # 시각화를 위해 matplotlib 불러오기

G_0 = nx.Graph() # 비어 있는 그래프를 만든다.

G_0.add_nodes_from(('a','b','c','d','e')) # 비어있는 그래프에 노드 추가하기
G_0.add_edges_from([('a','b'), ('a','c'),('a','d'),('a','e'),('d','e'),('c','e'),('b','a'),('e','a')]) #비어 있는 그래프에 edge 추가

nx.draw(G_0) # 기본형태의 네트워크 구조







```

```python


 ### 2.네트워크 구조에 nodes 이름 입히기
    


G = nx.Graph() # 비어 있는 그래프를 만든다.

G.add_nodes_from(('a','b','c','d','e')) # 비어있는 그래프에 노드 추가하기
G.add_edges_from([('a','b'), ('a','c'),('a','d'),('a','e'),('d','e'),('c','e'),('b','a'),('e','a')]) #비어 있는 그래프에 edge 추가

nx.draw(G, with_labels=True)

```

```python
### 3. 노드의 크기 조절 및 색깔 입히기

G_1 = nx.Graph() # 비어 있는 그래프를 만든다.

G_1.add_nodes_from(('a','b','c','d','e')) # 비어있는 그래프에 노드 추가하기
G_1.add_edges_from([('a','b'), ('a','c'),('a','d'),('a','e'),('d','e'),('c','e'),('b','a'),('e','a')]) #비어 있는 그래프에 edge 추가

nx.draw(G_1, with_labels=True, font_color="yellow", node_size=1000)
```

```python
### 3. 방향성 있는 네트워크 구조 만들기

G_2 = nx.DiGraph() # 방향성 있는 빈 그래프 생성

G_2.add_nodes_from(('a','b','c','d','e')) # 비어있는 그래프에 노드 추가하기
G_2.add_edges_from([('a','b'), ('a','c'),('a','d'),('a','e'),('d','e'),('c','e'),('b','a'),('e','a')]) #비어 있는 그래프에 edge 추가

nx.draw(G_2, with_labels=True, font_color='black' )
```

```python
### 실습예제1

import pandas as pd
import numpy  as np

feat = pd.read_csv("featuring.csv") #피처링 관계 데이터 불러오기
feat
```

```python
feat_iu = (feat['from']== '아이유')  | (feat['to']== '아이유') #아이유가 피처링 관련 데이터만 추출하기
```

```python
iu= feat[feat_iu] # 한글인식이 어려움으로 한글을 영어로 수정
iu = iu.replace({'from':'에픽하이'},{'from':'Epik High'})
iu = iu.replace({'from':'아이유'},{'from':'IU'})
iu =iu.replace({'to':'아이유'},{'to':'IU'})
iu =iu.replace({'to':'오혁'},{'to':'OHHYUK'})

iu
```

```python
iu2= np.array(iu) #데이터프레임을 리스트로 변경

```

```python


G_iu = nx.DiGraph() # 방향성 있는 빈 그래프 생성

G_iu.add_nodes_from([(iu2[0,0],iu2[0,1],iu2[1,0],iu2[2,1],iu2[3,0])]) # 비어있는 그래프에 아이유 데이터 추가
G_iu.add_edges_from([(iu2[0,0],iu2[0,1]),(iu2[1,0],iu2[0,0]),(iu2[0,0],iu2[2,1]),(iu2[3,0],iu2[0,0])]) #비어 있는 그래프에 edge 추가

nx.draw(G_iu, with_labels=True, font_color='red', node_size=950)
```
