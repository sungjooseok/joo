### bag of words  
#### (1)vector space model
- 문서를 bag of words로 표현  
- 단어가 쓰여지는 순서는 무시  
- 모든 문서에 한번 이상 나타난 단어들에 대해 유 (1)/무(0) 로 문서를 표현


```python
from konlpy.tag import Okt
import re  
okt=Okt()  

pra2=re.sub("(\.)","",pra)  # 온점을 제거

pra2=okt.morphs(pra2) # OKT 형태소 분석기를 통해 토큰화 작업을 수행


word2index={}  
bow=[]  
for voca in pra2:  
         if voca not in word2index.keys():  
             word2index[voca]=len(word2index)  # pra2을 읽으면서, word2index에 없는 (not in) 단어는 새로 추가하고, 이미 있는 단어는 넘김   
             bow.insert(len(word2index)-1,1)

         else:
            index=word2index.get(voca)# 재등장하는 단어의 인덱스를 받아옴

            bow[index]=bow[index]+1 # 단어의 갯수 세기
print(word2index)  
```

### TFIDF  
- 단어의 count를 단어가 나타난 문서의 수로 나눠서 자주 등장하지 않는 단어의 weight를 올림  
- tf(d, t): 문서 d에 단어 t가 나타난 횟수, count vector와 동일, 로그스케일 등 다양한 변형이 있음  
- df(t): 전체 문서 중에서 단어 t를 포함하는 문서의 수  
- idf(t): df(t)의 역수를 바로 쓸 수도 있으나, 여러가지 이유로 로그스케일과 스무딩을 적용한 공식을 사용,log(n/(1+df(t)), n은 전체 문서 수

```python
import pandas as pd 
from math import log 
```

```python
docs = [
  '재미있는 엘리스',
  '재미있는 판타지 드라마 ',
  '드라마 중에 제일 제미있는 엘리스',
  '엘리스가 재미져',
    '드라마가 좋더라'
] 
vocab = list(set(w for doc in docs for w in doc.split()))
vocab.sort()
```

```python
N = len(docs) # 총 문서의 수

def tf(t, d):
    return d.count(t)

def idf(t):
    df = 0
    for doc in docs:
        df += t in doc
    return log(N/(df + 1))

def tfidf(t, d):
    return tf(t,d)* idf(t)
```

```python
result = []
for i in range(N): # 각 문서에 대해서 아래 명령을 수행
    result.append([])
    d = docs[i]
    for j in range(len(docs)):
        t = docs[j]        
        result[-1].append(tf(t, d))

tf_ = pd.DataFrame(result, columns = docs)
tf_result = []
for i in range(N): # 각 문서에 대해서 아래 명령을 수행
    result.append([])
    d = docs[i]
    for j in range(len(vocab)):
        t = vocab[j]        
        result[-1].append(tf(t, d))

tf_ = pd.DataFrame(result, columns = vocab)
tf_
```
